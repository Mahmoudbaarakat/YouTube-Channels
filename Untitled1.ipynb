# Libraries

```python
from googleapiclient.discovery import build
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

# Create a YouTube API client and Retrieve the most popular videos in US

```python
# Create a YouTube API client
youtube = build('youtube', 'v3', developerKey='*************')

# Example: Retrieve the most popular videos in the US
response_videos = youtube.videos().list(
    part='snippet,contentDetails,statistics',
    chart='mostPopular',
    regionCode='US',
    maxResults=500  
).execute()

video_data = []

# Extract relevant information from each video
for video in response_videos['items']:
    title = video['snippet']['title']
    views = video['statistics']['viewCount']
    likes = video['statistics'].get('likeCount', 0)
    dislikes = video['statistics'].get('dislikeCount', 0)
    comments = video['statistics'].get('commentCount', 0)
    duration = video['contentDetails']['duration']
    published_at = video['snippet']['publishedAt']
    category_id = video['snippet']['categoryId']

   
    published_at_timestamp = pd.to_datetime(published_at)

    
    response_categories = youtube.videoCategories().list(
        part='snippet',
        id=category_id
    ).execute()

    category_name = response_categories['items'][0]['snippet']['title']

    video_data.append({
        'Title': title,
        'Views': views,
        'Likes': likes,
        'Dislikes': dislikes,
        'Comments': comments,
        'Duration': duration,
        'PublishedAt': published_at_timestamp,
        'CategoryID': category_id,
        'CategoryName': category_name
    })

# Create a DataFrame from the list of video data
df = pd.DataFrame(video_data)
```

```python
df.head(5)
```

```python
import re
df['#Comments'] = pd.to_numeric(df['Comments'], errors='coerce')
df['#Likes'] = pd.to_numeric(df['Likes'], errors='coerce')
df['#Views'] = pd.to_numeric(df['Views'], errors='coerce')
df['#Duration'] = pd.to_numeric(df['Duration'], errors='coerce')
df['minutes'] = df['Duration'].apply(lambda x: int(re.search(r'PT(\d+)M', x).group(1)) if re.search(r'PT(\d+)M', x) else 0)
df['seconds'] = df['Duration'].apply(lambda x: int(re.search(r'PT(\d+)S', x).group(1)) if re.search(r'PT(\d+)S', x) else 0)
df['#Duration'] = df['minutes'] * 60 + df['seconds']
```

# Data Analysis

```python
#Distribution of Videos Across Categories

category_counts = df['CategoryName'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Videos Across Categories')
plt.show()
```

```python
#Top 10 Videos by Views

top_videos_views = df.sort_values(by='#Views', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='#Views', y='Title', data=top_videos_views, palette='viridis')
plt.title('Top 10 Videos by Views')
plt.xlabel('Views')
plt.ylabel('Video Title')
plt.show()
```

```python
#Likes VS Comments based on Views

plt.figure(figsize=(10, 8))
sns.scatterplot(x='#Likes', y='#Comments', data=df, hue='#Views', palette='plasma', size='#Views')
plt.title('Likes VS Comments')
plt.xlabel('Likes')
plt.ylabel('Comments')
plt.show()
```

```python
#Correlation between views, likes, comments and duration

numeric_df = df.drop(['Title','Views','Duration', 'PublishedAt','Comments','Duration','minutes','seconds','CategoryID','CategoryName'], axis=1)
correlation_matrix = numeric_df.corr()
correlation_matrix['#Views']
```

```python
#Video duration analysis

plt.figure(figsize=(10, 6))
sns.histplot(df['#Duration'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Distribution of Video Durations')
plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Distribution of Engagement Rates by Category


df['EngagementRate'] = (df['#Likes'] + df['#Comments']) / df['#Views']
plt.figure(figsize=(12, 8))
sns.violinplot(x='CategoryName', y='EngagementRate', data=df, palette='viridis', inner='quartile')
plt.title('Distribution of Engagement Rates by Category')
plt.xlabel('Category')
plt.ylabel('Engagement Rate')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

```python
#The effect of video duration on engagement metrics.

# Scatter plot of Duration vs. Views
plt.figure(figsize=(8, 4))
sns.scatterplot(x='#Duration', y='#Views', data=df)
plt.title('Scatter Plot of Duration vs Views')
plt.xlabel('Duration (seconds)')
plt.ylabel('Views')
plt.show()

# Scatter plot of Duration vs. Likes
plt.figure(figsize=(8, 4))
sns.scatterplot(x='#Duration', y='#Likes', data=df)
plt.title('Scatter Plot of Duration vs. Likes')
plt.xlabel('Duration (seconds)')
plt.ylabel('Likes')
plt.show()

# Scatter plot of Duration vs. Comments
plt.figure(figsize=(8, 4))
sns.scatterplot(x='#Duration', y='#Comments', data=df)
plt.title('Scatter Plot of Duration vs. Comments')
plt.xlabel('Duration (seconds)')
plt.ylabel('Comments')
plt.show()
```
